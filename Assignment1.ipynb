```
weight = float(input("Enter weight in pounds:"))
height =  float(input("Enter height in inches:"))
w = weight*0.45359237
#print(w)
h = height*0.0254
#print(h)
BMI =w/(h*h)
#print(BMI)
if BMI <18.5:
  print("Underweight")
elif (BMI >=18.5) and (BMI < 25.0):
  print("Normal")
elif (BMI >=25.0) and (BMI<30.0):
  print("Overweight")
else:
  if BMI>=30.0:
    print("Obese")

```
